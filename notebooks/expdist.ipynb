```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
lam = 0.5
y = lam * np.exp(-lam * x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('Exponential Distribution')
plt.title('Exponential Distribution with lambda = 0.5')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 100, 100)
lambda_values = np.arange(10.0, 40.0, 5.0)

for lam in lambda_values:
    y = lam * np.exp(-lam * 1/(lam**1.5)*x) + 1
    plt.plot(x, y, label=f'lambda = {lam}')

plt.xlabel('x')
plt.ylabel('Exponential Distribution')
plt.title('Exponential Distributions with Varying Lambda')
plt.legend()
plt.grid(True)
plt.show()
```

By setting

$\sigma = 1-(t/t_I)$

Where $t$ is the number of tokens in the vector, and extra weight should be given to the first $t_I$ tokens, $\sigma$ is a coefficient that can be used in the exponential distribution equation, we can set the horizontal expansion factor to be $\mu = \frac{1}{\lambda^\sigma}$ to expand the graph horizontally, thus achieving the desired affect of increasing the weight of the 

We can use the equation $y = \lambda^{\frac{-\lambda x}{\lambda^{\sigma}}} + 1$ as our probability distribution funciton. This PDF will be used to increase the weight of the tokens at the beginning of the tokenized HTML text which represent such things like the title on the window and H1 headers.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 100, 100)

tokens = 500
interesting_tokens_list = np.arange(10.0, 50.0, 5.0)
for interesting_tokens in interesting_tokens_list:
    lam = tokens/interesting_tokens
    y = lam * np.exp(-lam * 1/(lam*lam**(1-interesting_tokens/tokens))*x) + 1
    plt.plot(x, y, label=f'it = {interesting_tokens}')

plt.xlabel('x')
plt.ylabel('Exponential Distribution')
plt.title('Exponential Distributions with Varying Lambda')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 200, 100)
interesting_tokens_list = np.arange(10.0, 41.0, 5.0)

for tokens in range(300, 1001, 100):
    plt.figure()  # Create a new figure for each value of "tokens"
    for interesting_tokens in interesting_tokens_list:
        lam = tokens / (interesting_tokens)
        coefficient = 1 - 1/(interesting_tokens)
        y = lam * np.exp(-lam * 1/(lam*lam**coefficient)*x) + 1
        plt.plot(x, y, label=f'tokens = {tokens}, it = {interesting_tokens}')

    plt.xlabel('x')
    plt.ylabel('Exponential Distribution')
    plt.title(f'Exponential Distributions with Tokens = {tokens}')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 100, 100)

factor = 0.1
for tokens in range(300, 1001, 100):
    interesting_tokens = tokens * factor
    lam = tokens / (interesting_tokens)
    coefficient = 1 - 1/(interesting_tokens)
    y = lam * np.exp(-lam * 1/(lam*lam**coefficient)*x) + 1
    plt.plot(x, y, label=f'tokens = {tokens}')

plt.xlabel('x')
plt.ylabel('Exponential Distribution')
plt.title(f'Exponential Distributions with coefficient {factor}')
plt.legend()
plt.grid(True)
plt.show()
```
